# Model Evaluation (5 min)
There are many metrics of model evaluation in solving Regression Problem. So far, we've learned one of the commonly used metrics, Mean Squared Error.
$$ \text{MSE}= \frac{1}{n}\sum^n_{i=1}(y^{(i)}-\hat y^{(i)})^2 $$
Intuitively, what Mean Squared Error does is to find the mean of the squared difference between the actual value and the predicted value(aka Error^2).

However, as the value generated is the squared difference, the error might not reflect the actual difference between the predicted value and actual value(aka lack of interpretability)

There is another metrics called Mean Absolute Error whereby it produces the mean of absolute error instead of squared error.
$$\text{MAE} = \frac{1}{n}\sum^n_{i=1}|y^{(i)}-\hat y^{(i)}|$$

Now let us implement MAE and observe the error produced.
