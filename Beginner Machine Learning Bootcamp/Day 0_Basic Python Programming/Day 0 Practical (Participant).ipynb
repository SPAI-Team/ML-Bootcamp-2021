## Car Plate Checksum

In general, every motor vehicle in Singapore requires a vehicle registration number to be displayed at the front center (in almost all cases or otherwise set by car manufacturer due to bumper constraints) and rear of the vehicle.
![](https://upload.wikimedia.org/wikipedia/commons/thumb/e/e8/Singapore_licence_plate_1990_rear_SDN7484U.jpg/450px-Singapore_licence_plate_1990_rear_SDN7484U.jpg)

All car plates have a unique number known as a **checksum**. It is a unique number which is calculated from the car plate number using a special algorithm. It is called a check sum because it involves **adding (summing)** up the numbers in the plate number in a special way, then using the final sum to **check the validity** of the overall number.

This checksum can be used to check if the car plate number is valid.
