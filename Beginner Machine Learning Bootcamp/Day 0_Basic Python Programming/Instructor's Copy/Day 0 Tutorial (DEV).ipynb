## Part 4.1 Arithmetic Operators

- $+$ : Add num1 and num2<br>
- $-$ : Subtract num2 from num1<br>
- / : Divide num1 by num2<br>
- $*$ : Multiply num1 by num2<br>
- % : Get the remainder after dividing num1 by num2<br>
- ** : num1 to the power of num2<br>
- // : Divide num1 by num3 and discard any decimal points from the answer<br>



```python
family_A_income = 7204.25
family_B_income = 4551
family_C_income = 10000

total_income = family_A_income + family_B_income + family_C_income 
average_income = total_income / 3

family_A_annual_income = family_A_income * 12
```

```python
# How do we extract the last digit?
student_id = 1941442
print(student_id % 10)
student_id //= 10
print(student_id)
print(student_id % 10)
```

## Part 4.2 Comparison Operators
- $==$ : If the values of two operands are equal, then the condition becomes true.
- $!=$ : 	If values of two operands are not equal, then condition becomes true.
- $<>$ : If values of two operands are not equal, then condition becomes true.
-$ > $: If the value of left operand is greater than the value of right operand, then condition becomes true.
- $<$ : If the value of left operand is less than the value of right operand, then condition becomes true.
- $>=$ : 	If the value of left operand is greater than or equal to the value of right operand, then condition becomes true.
- $<=$ :	If the value of left operand is less than or equal to the value of right operand, then condition becomes true.

```python
# When we want to compare the values of 2 numerical values, we need to use  == , = is used to assign values

password = input("Input a password: ")

access_granted = password == "abc"
access_granted
```

```python
max_capacity = 20
groups = 4
people_per_group = 2

max_capacity >= (groups * people_per_group)
```

## Part 6.1 : For Loop
- A for loop is used for iterating over a sequence 

For loop syntax: for stepper_variable in sequence_variable
   something_you_want_to_do


```python
jan_sales = [4000, 2441, 4314, 4144, 3165, 1553, 5313, 5313, 2341, 645, 4134, 3134, 5315, 3143, 3144, 1343, 1343, 4123, 1453, 1343, 2341, 4343, 5141, 1023, 2344, 3141, 4133, 4141, 1431, 424, 5135]
len(jan_sales)
```

```python
total_sales = 0
for sales_day in jan_sales:
  total_sales += sales_day
print(total_sales)
```

```python
for i in range(len(jan_sales)):
  print(f"Day {i + 1} Sales: {jan_sales[i]}")
```

```python
for i in range(9):
  if i % 2 != 0:
    print("Hello")
```

## Part 6.2 : While Loop
While loop repeatedly executes as long as a given condition is true

```python
population = 10
day = 0
while population < 20:
  day = day + 1
  population = int(population * 1.2)
print(day)
print(population)
```

## Part 6.3 : Infinite Loops

```python
while True :
    print("This is an infinite loop")
# an infinite loop runs on forvever because there is no condition to step out of the loop 
```

However, adding a break statement in a loop stops the loop from running infinitely. The break statement can be added to step out of a loop 

```python
while True:
    print("Breaking out of an infinite loop")
    break
```

## Part 7.1 : Defining a Function
Syntax : def function_name (parameter1,parameter2) :


```python
tweet_1 = {
    "Author" : "@hello_world",
    "Message" : "Wait a minute....",
    "Retweets" : 12,
    "Likers" : ["@mikey", "@akmal", "@goose"]
}

tweet_2 = {
    "Author" : "@akmal",
    "Message" : "Hello there!",
    "Retweets" : 32,
    "Likers" : ["@hello_world", "@yuna"]
}
def print_tweet(tweet): # defining a function
  print(tweet["Message"])
  print(f"- {tweet['Author']}")
  print(f"RTs: {tweet['Retweets']}; Likes: {len(tweet['Likers'])}")

print_tweet(tweet_1)
print_tweet(tweet_2)
```
