### Ordinal Encoder

To perform encoding for Ordinal Data, we can make use of `OrdinalEncoder` from sklearn library.

````python
df_copy["storey_range"].unique()
````

````python
from sklearn.preprocessing import OrdinalEncoder

enc_ord_storey = OrdinalEncoder()
enc_ord_storey.fit([['01 TO 03'],
                    ['01 TO 05'],
                    ['04 TO 06'],
                    ['06 TO 10'],
                    ['07 TO 09'],
                    ['10 TO 12'],
                    ['11 TO 15'],
                    ['13 TO 15'],
                    ['16 TO 18'],
                    ['16 TO 20'],
                    ['19 TO 21'],
                    ['21 TO 25'],
                    ['22 TO 24'],
                    ['25 TO 27'],
                    ['26 TO 30'],
                    ['28 TO 30'],
                    ['31 TO 33'],
                    ['34 TO 36'],
                    ['36 TO 40'],
                    ['37 TO 39'],
                    ['40 TO 42'],
                    ['43 TO 45'],
                    ['46 TO 48']
                    ])
storey_encoded = enc_ord_storey.transform(df_copy[["storey_range"]])
storey_encoded
````

````python
df_copy["storey_range"]
````

````python
df_copy["storey_range"] = storey_encoded
````
