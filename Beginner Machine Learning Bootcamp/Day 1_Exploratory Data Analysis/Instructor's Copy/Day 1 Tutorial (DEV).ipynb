## Introducing DataFrames and Series
There are three different types of data structures in pandas - Series, DataFrames, Panel. Throughout this course, we will only be going through mainly on DataFrame.

DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like an Excel spreadsheet or SQL table, or a dict of Series objects.

Series is a 1-dimensional labeled data structure, of the same data data type. An analogy would be Series is like a single column in an Excel speadsheet

## Data Visualisation

Data Visualisation is a form of visual art that grabs our interest and keeps our eyes on the message. When we see a chart, we quickly see trends and outliers. If we can see something, we internalize it quickly. It’s storytelling with a purpose. If you’ve ever stared at a massive spreadsheet of data and couldn’t see a trend, you know how much more effective a visualization can be.

image.png <br>
<i>Confirmed COVID-19 Cases built using Plotly visualisation library.</i>

Pandas has a set of built-in data visualisation features that provide some quick plots to access datasets. However, it may not be as detailed compared to other libraries, such as seaborn and plotly.

To begin plotting a chart, use the .plot followed by the desired chart.
For exmaple, say you want to plot a bar chart:
*df.plot.bar()*

[Back to the Content Table](#Top)

```python
pd.crosstab(titanic['sex'], titanic['survived'])
```

### Bar Chart
Bar Chart is used to compare between different categories (eg. gender, countries), oftenly used to rank and see the difference between categories. We use the .bar() method to create a bar chart.

```python
titanic['survived'].value_counts()
```

```python
titanic['survived'].value_counts().plot(kind='bar') # .value_counts() summarise the number of each value in the column
plt.show()
```

### Histogram
Histograms are used to determine the demographic of the data, meaning the spread and the concentration of the data for each class. We use the .hist() method to create a histogram

```python
titanic['age'].plot(kind='hist')
plt.show()
```

We can plot multiple columns into a histogram. Note that we use `.hist()` instead of `.plot(kind = 'hist')` here to show the distribution of each individual column instead of lumping all of them into one chart.

```python
titanic[['age', 'fare', 'sibsp', 'parch']].hist()
plt.show()
```

### Boxplot
Boxplot shows the maximum point, minimum point, and medium of the data. We use the .box() method to create a boxplot.

```python
titanic['age'].plot(kind='box', vert=False)
plt.show()
```

```python
titanic[['age', 'fare', 'sibsp', 'parch']].plot(figsize=(16, 9), kind='box', vert=False)

plt.show()
```

### Scatter Plot
We can use the .scatterplot() method for creating a scatterplot, and just as in Pandas we need to pass it the column names of the x and y data, but now we also need to pass the data as an additional argument because we aren’t calling the function on the data directly as we did in Pandas.

```python
titanic.plot(kind = 'scatter', y='fare', x='age')
plt.show()
```

### Correlation Plot
Correlation matrix indicates the strength and direction of a linear relationship between two variables.

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(), annot=True) #annot = True to include the numbers
```
