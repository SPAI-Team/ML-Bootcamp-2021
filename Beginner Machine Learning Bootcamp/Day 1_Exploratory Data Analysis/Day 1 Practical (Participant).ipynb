# EDA/Describing Data
Once we have imported our dataframe, we need to explore and understand what is inside the dataframe before we can build any model with it.

We can make use of simpler pandas method to generate text summaries or a more sophisticated libraries like pandas-profilling which takes longer to generate especially with a large dataframe.

Explore the imported dataframe with introduced methods:

1. `.head()` : Display first 5 rows by default. Can specify number of rows to display by parsing it as a parameter.
2. `.shape` : Display the shape of the dataframe in a form of tuple with following format: `(#rows, #columns)`
3. `.isnull().sum()` : Display the number of missing values in each column
4. `.dtypes` : Display the datatypes for each columns. Please Note that **each column can only have one datatype**(Unlike list and tuple introduced in Day 1).
5. `.info()` : Diplay summaries of dataframe including the shape, columns, datatype and null values
6. `.describe()` : Display statistical summaries for numerical columns.
